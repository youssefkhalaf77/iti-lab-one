````python
# Create a Python dictionary of 3 of your favourite foods with
# The keys of the dictionary should be the food name and the values their price
import matplotlib.pyplot as plt
%matplotlib inline

# 1. Create dictionary of favourite foods with prices
foods = {
    "Pizza": 8,
    "Burger": 5,
    "Sushi": 12
}

# Separate keys and values
food_names = list(foods.keys())
food_prices = list(foods.values())
````

Now let's try make some subplots. A subplot is another name for a figure with multiple plots on it.

````python
# Create an empty subplot with 2 rows and 2 columns (4 subplots total)
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

X = np.linspace(0, 10, 100)

# Setup 2 rows, 2 columns
fig, axes = plt.subplots(2, 2, figsize=(10,8))

# Top-left: line plot
axes[0,0].plot(X, np.sin(X), color="blue")
axes[0,0].set_title("Sine Wave")

# Top-right: line plot
axes[0,1].plot(X, np.cos(X), color="red")
axes[0,1].set_title("Cosine Wave")

# Bottom-left: scatter
axes[1,0].scatter(X, np.exp(X/3), color="green", alpha=0.6)
axes[1,0].set_title("Exponential Growth")

# Bottom-right: histogram
data = np.random.randn(500)
axes[1,1].hist(data, bins=20, color="purple", edgecolor="black")
axes[1,1].set_title("Normal Distribution Histogram")

plt.tight_layout()
plt.show()
````

````python
# Create the same plot as above with 2 rows and 2 columns and figsize of (10, 5)


# Plot X versus X/2 on the top left axes


# Plot a scatter plot of 10 random numbers on each axis on the top right subplot


# Plot a bar graph of the favourite food keys and values on the bottom left subplot


# Plot a histogram of 1000 random normally distributed numbers on the bottom right subplot
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Sample data
X = np.linspace(0, 10, 100)

# Favourite foods dictionary
foods = {"Pizza": 8, "Burger": 5, "Sushi": 12}
food_names = list(foods.keys())
food_prices = list(foods.values())

# Setup subplots (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(10, 5))

# 1. Top-left: Line plot X vs X/2
axes[0,0].plot(X, X/2, color="blue")
axes[0,0].set_title("X vs X/2")

# 2. Top-right: Scatter plot of 10 random numbers
rand_x = np.random.rand(10)
rand_y = np.random.rand(10)
axes[0,1].scatter(rand_x, rand_y, color="red", marker="o")
axes[0,1].set_title("Random Scatter (10 points)")

# 3. Bottom-left: Bar graph of favourite foods
axes[1,0].bar(food_names, food_prices, color="orange")
axes[1,0].set_title("Favourite Foods")
axes[1,0].set_ylabel("Price ($)")

# 4. Bottom-right: Histogram of normal distribution
data = np.random.randn(1000)
axes[1,1].hist(data, bins=30, color="green", edgecolor="black")
axes[1,1].set_title("Normal Distribution Histogram")

# Adjust layout
plt.tight_layout()
plt.show()
````

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

````python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#    
car_sales = pd.DataFrame({
    "Make": ["Toyota", "Honda", "Toyota", "BMW", "Nissan",
             "Toyota", "Honda", "Honda", "Toyota", "Nissan"],
    "Colour": ["White", "Red", "Blue", "Black", "White",
               "Green", "Grey", "White", "Blue", "Black"],
    "Odometer (KM)": [150043, 87899, 32549, 11179, 213095,
                      99213, 45698, 54738, 60000, 31600],
    "Doors": [4, 4, 3, 5, 4, 4, 4, 4, 4, 4],
    "Price": ["$4000", "$5000", "$7000", "$22000", "$3500",
              "$4500", "$7500", "$7000", "$6250", "$9700"]
})

# 1.    Price 
car_sales["Price"] = car_sales["Price"].str.replace(r'[\$,]', '', regex=True).astype(int)
````

````python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column (cumsum())


# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)

# View the car_sales DataFrame

# 2.   Total Sales (cumulative sum)
car_sales["Total Sales"] = car_sales["Price"].cumsum()

# 3.   Sale Date (   )
car_sales["Sale Date"] = pd.date_range(start=pd.Timestamp.today().normalize(), periods=len(car_sales), freq="D")

#  
print(car_sales)
````

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# ---   car_sales      ---
car_sales = pd.DataFrame({
    "Make": ["Toyota", "Honda", "Toyota", "BMW", "Nissan",
             "Toyota", "Honda", "Honda", "Toyota", "Nissan"],
    "Colour": ["White", "Red", "Blue", "Black", "White",
               "Green", "Grey", "White", "Blue", "Black"],
    "Odometer (KM)": [150043, 87899, 32549, 11179, 213095,
                      99213, 45698, 54738, 60000, 31600],
    "Doors": [4, 4, 3, 5, 4, 4, 4, 4, 4, 4],
    "Price": ["$4000", "$5000", "$7000", "$22000", "$3500",
              "$4500", "$7500", "$7000", "$6250", "$9700"]
})

#  Price 
car_sales["Price"] = car_sales["Price"].str.replace(r'[\$,]', '', regex=True).astype(int)

#  Total Sales cumulative
car_sales["Total Sales"] = car_sales["Price"].cumsum()

#    
car_sales["Sale Date"] = pd.date_range(start=pd.Timestamp.today().normalize(), periods=len(car_sales), freq="D")
````

````python
# Convert the 'Price' column to the integers
car_sales.plot(x="Sale Date", y="Total Sales", kind="line", figsize=(8,5), legend=False, marker="o")
plt.title("Total Sales Over Time")
plt.ylabel("Total Sales ($)")
plt.xlabel("Sale Date")
plt.show()

# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter", figsize=(7,5), color="red")
plt.title("Car Price vs Odometer")
plt.show()
````
